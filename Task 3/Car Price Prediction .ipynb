1) **how does cars price decides whether a customer wants to by that car or not?**
    * **Affordability:** Customers assess if the car's price aligns with their budget and financial capabilities. If the price is too high, it may be unaffordable and dissuade them from making a purchase.
    * **Value for Money:** Customers compare the price to the car's features, quality, and overall value proposition. A fair price for the perceived value is more likely to attract buyers.
    * **Financing Options:** The price affects the availability of financing options. A lower-priced car may offer more favorable financing terms, making it more appealing to customers who rely on loans.
    * **Resale Value:** Customers consider how the car's price will impact its future resale or trade-in value. A higher resale value can make a car more attractive, as it minimizes the long-term cost of ownership.
    * **Competition:** The price of a car is compared to similar models from other manufacturers. Customers may choose a car based on its competitive pricing relative to alternatives with similar features and capabilities.
<hr>
2) **What does a customer looks while buying a car?**
    * **Reliability and Durability:** Customers want a car that is known for reliability and durability. They look for brands and models with a track record of requiring fewer repairs and having a longer lifespan.
    * **Safety Features:** Safety is a top priority. Customers are drawn to cars with advanced safety features such as airbags, anti-lock brakes, collision avoidance systems, and high crash-test ratings.
    * **Fuel Efficiency:** With rising fuel costs and environmental concerns, customers are attracted to cars that offer good fuel efficiency. Vehicles with hybrid or electric options are increasingly popular for their eco-friendliness.
    * **Technology and Connectivity:** Modern customers expect advanced technology and connectivity features. Cars with intuitive infotainment systems, smartphone integration, and driver assistance technology (e.g., adaptive cruise control, lane-keeping assist) are highly attractive.
    * **Comfort and Interior Quality:** Interior comfort, quality of materials, and ergonomic design matter to customers. A well-designed, comfortable interior with spacious seating and user-friendly controls can be a significant selling point.

### Data preprocessing 

```python
del df['car_ID']
# since not relation table is present from the dataset hence car_id is being deleted
```

```python
for i in df.columns:
    print(i)
    print(df[i].unique())
    print(df[i].nunique())
    print('-'*50)
```

```python
dct = {'four': 4 , 'six': 6, 'five': 5 , 'three': 3, 'twelve':12, 'two':2, 'eight':8}
new = []
for i in df['cylindernumber']:
    if (i) in dct:
        new.append(dct[i])
df['cylindernumber'] = new
```

```python
# creating different columns for different cars
brand = []
model = []
for i in df.values:
    if len(i[1].split(' ')) >= 2:
        brand.append(i[1].split(' ')[0].lower())
        model.append(' '.join(i[1].split(' ')[1:]))
    else:
        brand.append(i[1])
        model.append('Missing')
        
df['Brand'] = brand
df['Model'] =  model
del df['CarName']
```

```python
# Outliers in a particular column
upper_limit = df['citympg'].mean() + 3*df['citympg'].std()
lower_limit = df['citympg'].mean() - 3*df['citympg'].std()
df[(df['citympg'] > upper_limit) | (df['citympg'] < lower_limit)]
```

```python
# number of outliers in each column
num = ['highwaympg','citympg','peakrpm','horsepower',
       'stroke','boreratio','curbweight','carheight',
       'carwidth','carlength','wheelbase']

for col in num:
    upper_limit = df[col].mean() + 3 * df[col].std()
    lower_limit = df[col].mean() - 3 * df[col].std()
    df[col] = np.where(df[col] > upper_limit,
                                upper_limit,
                                np.where(df[col] < lower_limit,
                                         lower_limit,
                                         df[col]))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, mean_absolute_error,precision_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
trns = ColumnTransformer(transformers=[('OHE',OneHotEncoder(),['fueltype','aspiration','doornumber',
                                                               'carbody','drivewheel','enginelocation',
                                                               'enginetype','fuelsystem'])],
                         remainder = 'passthrough' )
```

```python
X_train_trans = trns.fit_transform(X_train)
X_test_trans = trns.fit_transform(X_test)
```
